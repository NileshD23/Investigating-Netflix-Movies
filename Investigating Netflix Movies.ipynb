### 1. Loading our data into a dictionary
We have a CSV file containing Netflix data. We know that for the years 2011 to 2020, the average movie durations are 103, 101, 99, 100, 100, 95, 95, 96, 93, and 90, respectively.

````python
# view the netflix dataset
import pandas as pd

data = pd.read_csv('netflix_data.csv')
data.head()
````

````python
# create the years and durations lists
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

# create a dictionary with the years and durations
movie_dict = {'years': years, 'durations': durations}

# print the dictionary
movie_dict
````

### 2. Creating a DataFrame from a dictionary
Now convert the movie_dict to a DataFrame.

````python
# create a DataFrame from the dictionary
durations_df = pd.DataFrame(movie_dict)

# print the DataFrame
durations_df
````

### 3. A visual inspection of the data
Because the data is continuous, we can use a line plot to visualize the data.

````python
# import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
fig = plt.figure()

# draw a line plot of release_years and durations
plt.plot(durations_df['years'], durations_df['durations'])

# create a title
plt.title('Netflix Movie Durations 2011-2020')

# show the plot
plt.show()
````

### 4. Loading the rest of the data from the CSV
We still need to answer the following questions:
1. What does this trend lookl ike over a longer period of time?
2. Is this explainable by something like the genre of entertainment?
We will create a new DataFrame using all of the data from ```netflix_data.csv```. 

````python
# read in the CSV as a DataFrame
netflix_df = pd.read_csv('netflix_data.csv')

# print the first five rows of the DataFrame
netflix_df[0:5]
````

### 5. Filtering for movies
Create a new DataFrame ```netflix_movies``` containing only ```title, country, genre, release_year```, and ```duration```.

````python
# subset the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

# select only the oclumns of interest
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

# print the first five rows of the new DataFrame
netflix_movies_col_subset[0:5]
````

### 6. Creating a scatter plot
Now we should try to visualize the data over a longer range of time. Since we are working with aggregates, we'll make a scatter plot. 

````python
# create a figure and increase the figure size
fig = plt.figure(figsize=(12, 8))

# create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"])

# create a title
plt.title("Movie Duration by Year of Release")

# show the plot
plt.show()
````

### 7. Digging deeper
Some of the movies have a ```duration``` under 60 mins. We can filter these movies out.

````python
# filter for durations shorter than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]

# print the first 20 rows of short_movies
short_movies[0:20]
````

### 8. Marking non-feature films
We will now plot the data and mark the shorter films with a different color.

````python
# define an empty list
colors = []

# iterate over rows of netlfix_movies_col_subset
for lab, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre'] == 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')

# inspec the first 10 values in our list
colors[0:10]
````

### 9. Plotting with color
We'll add some details to our plot for visual flair. 

````python
# set the figure style and initialize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12, 8))

# create a scatter plot of duration versus release_year
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"], c=colors)

# create a title and axis labels
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release Year")
plt.ylabel("Duration (minutes)")

# show the plot
plt.show()
````
